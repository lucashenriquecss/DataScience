```python
import pandas as pd
```

```python
dados = pd.read_csv("aluguel.csv", sep=';')
```

```python
dados.head()
```

```python
#Selecione somente os imóveis classificados com tipo 'Apartamento'.
selecao = dados ['Tipo'] == 'Apartamento'
n1 = dados[selecao].shape[0]
n1
```

```python
#Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.
selecao = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
n2 = dados[selecao].shape[0]
n2
```

```python
#Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.
selecao = (dados['Area'] >= 60) & (dados['Area'] <= 100)
n3 = dados[selecao].shape[0]
n3
```

```python
#Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.
selecao = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
n4 = dados[selecao].shape[0]
n4
```

```python
print("Nº de imóveis classificados com tipo 'Apartamento' -> {}".format(n1))
print("Nº de imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'-> {}".format(n2))
print("Nº de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites -> {}".format(n3))
print("Nº de imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00 -> {}".format(n4))
```

##Relatorio de analise VI - Criando novas variaveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel.csv', sep =';')
dados.head()
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU'] #valor a ser pago todo mês
```

```python
dados.head(10)
```

```python
dados['Valor m2'] = dados['Valor']/ dados['Area']
dados.head(10)
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2) #arredondamento 
dados.head(10)
```

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto']/ dados['Area']).round(2)
```

```python
dados.head(10)
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Case de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x : 'Casa' if x in casa else 'Apartamento')#aplicando função a cada registro do dataframe
```

```python
dados.head(50)
```

# **Relatório de Análise VII**
Criando Agrupamentos

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel.csv', sep = ';')
dados.head(10)
```

```python
dados['Valor'].mean() #buscando a media
```

buscando a media dos valores por bairro

```python
bairros=['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
selecao = dados['Bairro'].isin(bairros)
dados=dados[selecao]
```

```python
dados['Bairro'].drop_duplicates()#removendo duplicidades
```

```python
grupo_bairro = dados.groupby('Bairro')
```

```python
for bairro, dados in grupo_bairro:
  print(f'{bairro} -> {dados.Valor.mean()}')
```

```python
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

ESTATISTICA DESCRITIVA

```python
grupo_bairro['Valor'].describe().round(2)
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max']).rename(columns = {'min' : 'Mínimo', 'max': 'Máximo'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (10,10))
```

```python
fig  = grupo_bairro['Valor'].mean().plot.bar(color= 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do aluguel por bairro')
```
